```python
import matplotlib.pyplot as plt
import random
```

```python
# Definir los vértices del triángulo
v1 = [0, 0]      # Vértice inferior izquierdo
v2 = [1, 0]      # Vértice inferior derecho
v3 = [0.5, 0.866]  # Vértice superior (altura de un triángulo equilátero)

# Lista para almacenar los puntos del fractal
points_x = []
points_y = []

# Inicializar un punto inicial aleatorio dentro del triángulo
current_point = [random.uniform(0, 1), random.uniform(0, 0.866)]

# Generar puntos mediante el método del "juego del caos"
num_points = 10000  # Número de puntos a generar

for _ in range(num_points):
    # Elegir aleatoriamente uno de los vértices
    chosen_vertex = random.choice([v1, v2, v3])
    # Mover el punto a la mitad de la distancia entre el punto actual y el vértice elegido
    current_point = [
        (current_point[0] + chosen_vertex[0]) / 2,
        (current_point[1] + chosen_vertex[1]) / 2
    ]
    # Agregar el punto actual a la lista
    points_x.append(current_point[0])
    points_y.append(current_point[1])

# Graficar el triángulo de Sierpiński
plt.figure(figsize=(8, 8))
plt.scatter(points_x, points_y, s=0.5, color="blue")
plt.axis("off")
plt.title("Triángulo de Sierpiński")
plt.show()
```
