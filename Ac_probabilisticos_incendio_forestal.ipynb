```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones del bosque
n = 50
bosque = np.random.choice([0, 1], size=(n, n), p=[0.7, 0.3])  # 0: vaco, 1: rbol sano
bosque[n//2, n//2] = 2  # 2: en llamas (foco inicial)

# Reglas del autmata
def propagar_fuego(bosque, p=0.3):
    nuevo_bosque = bosque.copy()
    for i in range(1, n-1):
        for j in range(1, n-1):
            if bosque[i, j] == 1:  # Si es un rbol sano
                vecinos_en_llamas = (bosque[i-1:i+2, j-1:j+2] == 2).sum()
                if vecinos_en_llamas > 0 and np.random.rand() < p:
                    nuevo_bosque[i, j] = 2
            elif bosque[i, j] == 2:  # Si est en llamas
                nuevo_bosque[i, j] = 3  # 3: quemado
    return nuevo_bosque

# Simulacin
for paso in range(20):
    plt.imshow(bosque, cmap="hot")
    plt.title(f"Paso {paso}")
    plt.show()
    bosque = propagar_fuego(bosque)
```
